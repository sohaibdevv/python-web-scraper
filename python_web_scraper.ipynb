```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# 1. Setup URL and Headers (to look like a real browser)
url = "https://en.wikipedia.org/wiki/List_of_brightest_stars"
headers = {'User-Agent': 'Mozilla/5.0'}

try:
    # 2. Send HTTP Request
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        # 3. Parse HTML
        soup = BeautifulSoup(response.content, 'html.parser')
        print(f"Successfully parsed: {soup.title.text}")

        # 4. Identify and Extract Data (Targeting the first table)
        star_data = []
        table = soup.find('table', {'class': 'wikitable'})
        rows = table.find_all('tr')

        for row in rows[1:]:  # Skip header
            # Use both 'td' and 'th' to account for possible row headers in data rows
            cols = row.find_all(['td', 'th'])
            # Ensure there are enough columns to access index 3
            if len(cols) > 3:
                # Handle missing data with checks
                name = cols[1].text.strip() if cols[1] else "N/A"
                magnitude = cols[3].text.strip() if cols[3] else "N/A"

                star_data.append({
                    "Star Name": name,
                    "Magnitude": magnitude
                })

        # 5. Create DataFrame
        df = pd.DataFrame(star_data)

        # 6. Save to CSV
        df.to_csv('brightest_stars.csv', index=False)
        print("Scraping complete! File saved as brightest_stars.csv")
        print(df.head())

    else:
        print(f"Failed to retrieve page. Status code: {response.status_code}")

except Exception as e:
    print(f"An error occurred: {e}")
```
